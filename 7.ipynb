**Алгоритм Работы**:
1. При помощи метода `.mean()` вычилсяется среднее значение столбца таблицы `["reading score"]`
2. Вычисляются абитуриенты, чьи баллы ниже среднего, а после суммируется их количество
3. Выводится результат

#**Задание 7.2**: Шифровальщик

**Алгоритм работы**:
1. Создается функция `crypter` что:
   
   1.1. Получает длину сообщения
   
   1.2. Возвращется сумма срезов после среднего символа + до среднего символа
3. в таблице создается столбец `'crypter'`, в который по средствам использование лямбды генерируется шифрованный столбец `'test preparation course'`
4. выводится результат первых 3х тсрок таблицы(в 2 столбца)

````python
#Данный код не работает без первого отрезка кода(см. выше)
def crypter(string):
    midLen = len(string)//2
    return string[midLen:] + string[:midLen]

data['crypted'] = data['test preparation course'].map(lambda dd: crypter(str(dd)))
print(f"Результат шифровки: \n{data[['test preparation course','crypted']].head()}")
````

# **Ответы на вопросы**:
1. Определение функции. Структура. Аргументы функции
Функция — это блок кода, выполняющий определённую задачу и возвращающий результат.
Синтаксис:
```
def имя_функции(аргументы):

    тело_функции
   
    return результат
```

Пример:
```
def add(a, b):

    return a + b

```
2. Вызов функции
Функцию можно вызвать по имени с передачей аргументов:
`
add(2, 3)  # вернёт 5

`
3. Функция как объект
В Python функция является объектом, её можно присвоить переменной, передавать в другую функцию или хранить в списке.
Пример:
```
f = add
f(1,2)  # вернёт 3
```

4. Функция map()
Применяет функцию ко всем элементам последовательности.
Синтаксис: `map(функция, итерируемый_объект)`
Пример:
```
list(map(lambda x: x*2, [1,2,3]))  # [2,4,6]
```

5. Лямбда-функции
Короткая анонимная функция, используется для одноразовых операций.
Пример: `lambda x: x**2` — возводит число в квадрат.

6. Функция `filter()`
Отбирает элементы последовательности по условию.
Пример:

`list(filter(lambda x: x>2, [1,2,3,4]))  # [3,4]`


7. Строка — итерируемый объект. Индексация элементов строки
Строка состоит из символов, к которым можно обращаться по индексу: строка[индекс].
Пример:
```
s = "Python"
s[0]  # 'P'
s[-1] # 'n'
```

8. Поиск подстроки и срез
Срез: `строка[начало:конец:шаг]` — получает часть строки.
Поиск подстроки: строка.find('подстрока') — возвращает индекс или -1, если нет.
Пример:
```
s = "Hello World"
s[0:5]  # 'Hello'
s.find("World")  # 6
```

9. Метод count()
Считает количество вхождений подстроки в строку.
Синтаксис: строка.count('подстрока')
Пример: `"banana".count('a')` → 3

10. Методы lower() и upper()
Преобразуют строку в нижний или верхний регистр.
Примеры:
```
"Python".lower()  # 'python'
"Python".upper()  # 'PYTHON'
```

11. Метод replace()
Заменяет одну подстроку на другую.
Синтаксис: строка.replace(старое, новое)
Пример: `"Hello".replace("H","J")` → `'Jello'`

12. Регулярные выражения. Модуль re
Используются для поиска и обработки строк по шаблону.
Примеры функций: re.search(), re.findall(), re.sub()
Синтаксис:
```
import re
re.findall(r'\d+', 'abc123')  # ['123']
```

CSV (Comma-Separated Values) — текстовый файл для хранения табличных данных, где значения разделены запятыми.
