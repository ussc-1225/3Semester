**Задание 4.1**: Работа со словарями

**Алгоритм работы**:
1. при помощи `input` спарсировать данные у пользователя
2. Разбить данеые на массив слов
3. Работает цикл перебора слов.
   3.1. слово проверяется по словарю определителях, и если подходит конец цикла
4. Выводится результат

**Ход работы**:

1. Вводится запрос
2. Обрабатывается запрос
3. Выводится результат



```python
phones = {
    "единый": "112",
    "пожар": "101",
    "полиция": "102",
    "скорая": "103",
    "газ": "104"
}
data = input("Опишите проблему: ")
for i in data.split():
    if i in phones.keys():
        print(f"подходящий номер: {phones[i]}")
        break
```

**Задание 4.2**: Подсчет количества пользователей,что попадали под события с категорией Table 1 раз.
** Алгоритм работы**:
1. При помощи библиотеки `json` загрущить файл по пути ./Data/data.json
2. Создать `DefaultDict` для подсчета кол-ва юзеров
3. при помощи цикла собрать все события в `DefaultDict`, что были у юзеров
4. при помощи выражения отсеять тех, кто применял это событие только 1 раз
5. при помощи метода `len()` подсчитать количество юзеров
6. Выводить всё это по мере появления нужных данных

```python
import collections, json
with open("./Data/data.json", "r") as file:
    data = json.load(file)
dataCounter = collections.defaultdict(int)
for i in data["events_data"]:
    if i["category"] == "page":
        dataCounter[i["user_id"]] += 1
finalData = [str(i) for i,k in dataCounter.items() if k == 1]
print(f" Итого 1 действие с page совершили юзеры: ")
for i in finalData: print("  " + i)
print("Суммарное кол-во юзеров:"+str(len(finalData)) )
```

**Ответы на вопросы**:
1. Что такое словарь
Словарь — это структура данных, хранящая пары «ключ — значение». Каждый ключ уникален и связан с определённым значением. Словари удобны для быстрого поиска данных по названию.
Пример: `{'имя': 'Артем', 'возраст': 20}`

2. Создание словаря
Синтаксис:
имя_словаря = {ключ: значение, ключ: значение}
Можно создать пустой словарь `{}` или заполненный заранее.
Пример: `student = {'name': 'Иван', 'grade': 4}`

3. Обращение к ключам словаря
К элементам словаря обращаются через квадратные скобки или метод get().
Примеры:
`student['name']` или `student.get('grade')`

4. Перебор элементов словаря

По ключам: `for k in dict.keys()`

По значениям: `for v in dict.values()`

По ключам и значениям одновременно: `for k, v in dict.items()`
Это позволяет проходить по всему содержимому словаря разными способами.

5. Вложенные списки
Вложенный список — это список, внутри которого находятся другие списки.
Пример: `[[1,2,3], [4,5,6], [7,8,9]]`
Используется для представления таблиц или многомерных данных.

6. Фильтрация вложенного списка
Можно отбирать элементы по условию, например оставить только строки, где сумма больше заданной.
Пример:` [row for row in data if sum(row) > 10]`
То есть отбор только тех подсписков, которые удовлетворяют условию.

7. Вложенные словари
Словарь, в котором значениями являются другие словари.
Пример:

`
students = {
    'Иван': {'оценка': 5, 'возраст': 18},
    'Анна': {'оценка': 4, 'возраст': 19}
}
`

8. Фильтрация вложенных словарей
Проверяется условие внутри каждого подсловаря, например оставить только студентов с оценкой выше 4:
`{k:v for k,v in students.items() if v['оценка'] > 4}`

9. JSON-файл. Назначение
JSON (JavaScript Object Notation) — формат хранения и обмена структурированными данными. Используется для передачи данных между программами и системами в виде текста.

10. Открытие JSON-файла в Python
Для работы используется модуль json.
Пример:
`
import json
with open('data.json', 'r', encoding='utf-8') as f:
    data = json.load(f)
`

После этого данные из файла превращаются в словари и списки Python.

11. Метод Counter
Метод Counter из модуля collections используется для подсчёта количества вхождений элементов.
Синтаксис:
`
from collections import Counter
Counter(последовательность)
`

Пример: `Counter(['a','b','a','c']) → {'a': 2, 'b': 1, 'c': 1}`
