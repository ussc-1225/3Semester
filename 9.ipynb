# **Задание 9.1**: Посдчет статитики чаевых

**Алгоритм работы**:
1. при помощи метода `pandas.read_csv` подгружается файл `"./Data/tips.csv"`
2. Посдчитываются в выводе всяческие параметры чаевыъ по типу Средних, медианы, моды и отклюнения тех же чаевый
3. создается `data.groupby` и попадает прямо в график
4. идут более корректные настройки данных по типу надписей, названий, legend а так-положения и стиль рисования графика
5. Показ итогового графика

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("./Data/tips.csv")

print(f"Среднее чаевых: {data['tip'].mean():.2f}")
print(f"Медиана чаевых: {data['tip'].median():.2f}")
print(f"Мода чаевых: {data['tip'].mode()[0]:.2f}")
print(f"Стандартное отклонение: {data['tip'].std():.2f}")
print(f"Q1: {data['tip'].quantile(0.25):.2f}, Q3: {data['tip'].quantile(0.75):.2f}, Межквартильный размах: {data['tip'].quantile(0.25) - data['tip'].quantile(0.75) :.2f}\n")


plt.figure(figsize=(10,6))

for smoker in data['smoker'].unique():
    subset = data.groupby(['day', 'smoker'])['tip'].mean().loc[:, smoker]
    subset.plot(kind='bar', label=f"{smoker}")

plt.title("Средние чаевые по дням и курению")
plt.xlabel("День недели")
plt.ylabel("Средние чаевые ($)")
plt.legend(title="Курильщик")
plt.grid(True, linestyle="--", alpha=0)
plt.show()
```
