**Задание 5.1**: работа с ЗП футболистов

**Алгоритм работы**:
1. При помощи метода `pandas.read_csv()` импортируется файл `"/Data/data.csv"`
2. Выбираются необходимые колоны таблицы для работы
3. Находится по началу люди с макс. зарплатой(`fileData[penalties  == penalties.max()]`), а после и со средней (`(fileData[penalties  == penalties.max()])['Wage'].mean()`)

```python
import pandas as pd

fileData = pd.read_csv("./Data/football.csv")
print("Коротко о таблице: \n\n ")
print(f"Первые 5 строк DataFrame: \n{fileData.head()}")
print(f"\n\nПоследние 5 строк DataFrame: \n{fileData.tail()}")
print(f"\n\nОбщая информация о DataFrame: \n{fileData.info()}")
print(f"\n\nСтатистика по числовым колонкам: {fileData.describe()}")
print(f"\n\nПример Series (первые 5 элементов): \n{fileData['Wage'].head()}")

print("\n\nЗадание: \n\n")
penalties = fileData['Penalties']
heading = fileData['HeadingAccuracy']

avgPenalties = (fileData[penalties  == penalties.max()])['Wage'].mean()
avgHeading = (fileData[heading == heading.max()])['Wage'].mean()
print(f"средняя зп с макс. пенальти: {avgPenalties} \nсредняя зп с макс точностью удара головой: {avgHeading} \nРазница в зп: {avgPenalties/avgHeading} раза")

```

#**Ответы на вопросы**:
1. Библиотека Pandas. Импорт библиотеки
Pandas — библиотека Python для работы с таблицами и структурированными данными. Позволяет удобно обрабатывать, анализировать и фильтровать данные.
Импорт выполняется так:
`import pandas`


2. Структура Series
Series — одномерный массив с метками индексов для каждого элемента. Может содержать числа, строки и другие типы данных.
Пример: `pd.Series([10, 20, 30])` создаёт серию с индексами 0,1,2 и значениями 10,20,30.

3. Создание Series
Синтаксис: `pd.Series(список/массив, index=список_индексов)`
Пример:

`s = pd.Series([100, 200, 300], index=['a','b','c'])`


Элементы можно обращаться по индексу: `s['b']` → 200

4. Функция display
Используется для красивого отображения объектов Pandas (Series или DataFrame) в Jupyter Notebook.
Синтаксис: display(объект)
Пример: `display(s)`

5. Доступ к элементам Series с использованием .loc или .iloc

`.loc` — доступ по метке индекса

`.iloc` — доступ по позиции элемента
Примеры:

`s.loc['a'] `  # значение с индексом 'a'
`s.iloc[0]`    # первый элемент


6. Объект DataFrame. Создание
DataFrame — двумерная таблица данных, строки и столбцы которой имеют метки.
Синтаксис создания:

`df = pd.DataFrame({'Имя':['Иван','Анна'], 'Возраст':[18,19]})`


Каждый столбец может быть Series или списком.

7. Функция read_csv
Позволяет загрузить CSV-файл в DataFrame.
Синтаксис:` pd.read_csv('имя_файла.csv')`
Пример: `df = pd.read_csv('data.csv')`

8. Получение информации о DataFrame: head и tail
`
df.head(n)` — показывает первые n строк

`df.tail(n)` — показывает последние n строк
Пример: `df.head(5)`

9. Получение информации о DataFrame: info
Выводит основную информацию: количество строк, типы данных, наличие пропусков.
Пример: `df.info()`

10. Получение информации о DataFrame: describe
Возвращает статистические характеристики числовых столбцов: среднее, стандартное отклонение, минимум, максимум, квартильные значения.
Синтаксис: `df.describe()`
Пример: `df.describe()`

11. Индексация и извлечение данных: статистические методы
С помощью методов Pandas можно вычислять статистику по столбцам или строкам:

`df.mean()` — среднее

`df.median() `— медиана

`df.max()` / `df.min()` — максимум/минимум

`df.sum()` — сумма
Пример: `df['Возраст'].mean()` → средний возраст в столбце
