# **Задание 8.1**: Подсчет самого популярного прибыльного жанра из файла `./Data/films.csv`

**Алгоритм работы**:
1. подключается файл `./Data/films.csv` при помощи `pandas.read_csv`
2. (опционально) Выводятся всевозможные данные о таблице внутри файла `./Data/films.csv`
3. Создается колонка `pribuil`, в котором подсчитывается прибыл для каждого фильма из таблицы
4. при помощи собственно написанного метода getGanres из таблицы парсируются все жанры, что подсчитываются при помощи `pandas.Series`
5. В выводе подсчитываются самый популярный жанр и самый популярный оттуда фильм

````python

import pandas as pd

data = pd.read_csv("./Data/films.csv")
'''
#Данные о таблице, решил не добавлять т.к. эти самые строки уже мелькали в 2-х лабах до этого, но если нужно можно просто убрать кавычки
print("Коротко о таблице: \n")
print(f"Первые 5 строк DataFrame: \n{data.head()}")
print(f"\n\nПоследние 5 строк DataFrame: \n{data.tail()}")
print(f"\n\nОбщая информация о DataFrame: \n{data.info()}")
print(f"\n\nСтатистика по числовым колонкам: \n{data.describe()}")
'''
data['pribuil'] = data['revenue'] - data['budget']

def getGanres(d):
    ganresList = []
    for i in d:
        if isinstance(i,str):
            ganresList += i.split('|')
    return ganresList

gC = pd.Series(getGanres(data[(data['pribuil'] > 0 )]["genres"])).value_counts()

print(f'Самый популярный жанр среди тех, у кого есть прибыль: {gC.index[0]}, со счетом в {gC.iloc[0]} фильмов')
````

# **Ответы на вопросы**:
1. Чтение файла. Функция read_csv()
Используется для чтения CSV-файлов в DataFrame.
Синтаксис:
`
pandas.read_csv(путь_к_файлу, sep=',', encoding='utf-8', header=0)
`

Пример:
```
import pandas as pd
data = pd.read_csv('sample.csv')
```

2. Функция columns()
Позволяет получить или изменить названия столбцов DataFrame.
Синтаксис:
```
data.columns        # получить список названий

data.columns = [...] # изменить
```

Пример:
`
data.columns = ['Имя', 'Возраст', 'Город']
`

3. Очистка данных. Валидность данных
Очистка данных — это процесс исправления или удаления некорректных, дублированных или отсутствующих значений.
Типовые шаги:

Проверка пропусков (`isna()`, `dropna()`, `fillna()`)

Проверка дубликатов (`duplicated()`)

Проверка типов данных (`dtypes`)

4. Функции unique() и info()
unique() — возвращает уникальные значения столбца.
info() — выводит типы данных и количество непустых значений.
Пример:
```
data['Город'].unique()
data.info()
```

5. Ошибки в данных (на примере sample.csv)
Типичные проблемы, встречающиеся в реальных данных:

5.1. Пропуски в столбцах (пустые ячейки)

5.2. Дубли строк

5.3. Некорректные типы (например, возраст строкой)

5.4. Лишние пробелы в строках

5.5. Опечатки и разные варианты написания (например, "Moscow", "moscow")

5.6. Невалидные значения (например, отрицательный возраст)

6. Удаление столбца. Функция drop()
Используется для удаления строк или столбцов.
Синтаксис:
`
data.drop('ИмяСтолбца', axis=1, inplace=True)
`

Пример:
`
data.drop('НенужныйСтолбец', axis=1, inplace=True)
`

7. Фильтрация ошибок. Функция query()
Позволяет фильтровать строки по условию.
Синтаксис:
`
data.query('Возраст > 18 and Город == "Москва"')
`

Пример:
`
взрослые = data.query('Возраст >= 18')
`

8. Функции str.match() и str.contains()
Используются для фильтрации строковых данных по шаблону.
Синтаксис и примеры:
```
data[data['Город'].str.match('Мо.*')]

        # начинается с 'Мо'
data[data['Имя'].str.contains('ив', case=False)]  # содержит 'ив'
```

10. Преобразование данных. Функция apply()

Синтаксис:

Пример:

`data['Возраст'] = data['Возраст'].apply(lambda x: int(x) if x != '' else 0)`
