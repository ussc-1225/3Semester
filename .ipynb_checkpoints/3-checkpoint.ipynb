**Задание 3.1** – Работа с текстом

**Описание задачи**:
Дан произвольный текст, содержащий слова и цифры. Необходимо сформулировать список из всех имеющихся в нем цифр, определите их количество, сумму и
найти максимальное.

**Ход работы**:

1. Пользователь вводит  текст.
2. Строка разбивается на список слов с помощью метода `split()`.
3. Используя функцию `isInt()` отсеять из текста НЕ числа.
4. Выводятся нужные данные

```python
def isInt(n):
    try: 
        b = int(n)
        return True
    except ValueError:
        return False
text = input("Текст: ")
textArr = text.split()
intInTextArr = [int(k) for k in textArr if isInt(k)]
print(f"Массив слов: {textArr}")
print(f"Числа из текста: {intInTextArr} \n Кол-во чисел: {len(intInTextArr)} \n Сумма число: {sum(intInTextArr)} \n Максимальное число: {max(intInTextArr)}")
```

# **Ответы на вопросы:**
1. Список
Список — это изменяемая последовательность объектов любого типа, объединённых в одну структуру. Он позволяет хранить числа, строки, другие списки и обращаться к элементам по индексу.
Пример: `[1, 2, 3, 'a', 'b']`

2. Создание списка
Синтаксис: `имя_списка = [элементы]`
Можно создавать пустой список: `[]`, или сразу со значениями: `[1, 2, 3]`.
Пример: `fruits = ['яблоко', 'банан', 'вишня']`

3. Метод split
Метод строк, который превращает строку в список слов по разделителю (по умолчанию — пробел).
Синтаксис: строка.split(разделитель)
Пример: `'Привет мир'.split() → ['Привет', 'мир']`

4. Индексация элементов списка
Элементы списка имеют порядковые номера (индексы), начиная с 0.
Можно обратиться к элементу через `список[индекс]` или с конца через отрицательные индексы.
Пример:

`fruits[0]  # 'яблоко'`

`fruits[-1] # 'вишня'`


5. Работа со срезами
Срез позволяет получить часть списка: `список[начало:конец:шаг]`

начало включается

конец не включается

шаг указывает через сколько элементов брать
Пример:

`
numbers = [1,2,3,4,5]`

`numbers[1:4]  # [2,3,4]`

numbers[::2]  # [1,3,5]`
`

6. Метод append
Добавляет элемент в конец списка.
Синтаксис: `список.append(элемент)`
Пример:

`
fruits.append('груша')  # ['яблоко','банан','вишня','груша']
`

7. Метод sort
Сортирует список по возрастанию (или убыванию с reverse=True).
Синтаксис: `список.sort()`
Пример:
`
numbers.sort()  # [1,2,3,4,5]`


`numbers.sort(reverse=True)  # [5,4,3,2,1]`
`

9. Функции min, max, метод count

`min(список)` — минимальный элемент

`max(список)` — максимальный элемент

`список.count(элемент)` — считает, сколько раз элемент встречается
Пример:
`
numbers = [1,2,2,3]`

`min(numbers)  # 1`

`max(numbers)  # 3`

`numbers.count(2)  # 2
`

9. Генераторы списков
Позволяют создавать списки компактно на основе других последовательностей с условием.
Синтаксис: `[выражение for переменная in последовательность if условие]`
Пример: `[x**2 for x in range(5)]` → `[0,1,4,9,16]`

10. Что делает данный код

`myList = [x for x in range(10, 100) if x%3 == 0 and x%5 == 0]`


Он создаёт список всех чисел от 10 до 99, которые одновременно делятся на 3 и на 5.
То есть это кратные 15 числа в указанном диапазоне: `[15,30,45,60,75,90]`.
