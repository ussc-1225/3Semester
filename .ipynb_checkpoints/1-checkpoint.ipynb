### Задание 1.1
Написать программу, которая:
- Принимает 2 числа от пользователя
- подставляет в формулу (x^y)^3 / (y^x)^2 + 0.7(xy)^1/3


```python
x = float(input("Число 1:"))
y = float(input("Число 2:"))
print(f"Вывод: {((x**(3*y)))/(y**(2*x)) + 0.7 * (x*y)**(1/3)} ")
```

### Задание 1.2
Написать программу, которая:
- Принимает число месяца (1-12)
- Выводит название месяца
- Определяет сезон (зима, весна, лето, осень) с использованием if/elif/else

```python
number = input("Короткий номер: ")
nums = {
    "112": "Единый номер",
    "101": "Пожарная охрана и МЧС",
    "102": "Полиция",
    "103": "Скорая помошь",
    "104": "аварийная газовая служба"
}
print(f"Код соотвествует: '{nums[number]}'")
```

ОТВЕТЫ НА ВОПРОСЫ:

Jupyter Notebook. Назначение Jupyter Notebook — это среда, где можно писать, запускать и сразу видеть результат кода.

Jupyter Notebook. Способы запуска Есть несколько способов: Через Anaconda Navigator → выбрать Jupyter Notebook → нажать Launch.

Notebook. Типы ячеек. Назначение. Создание В Jupyter есть два основных типа ячеек: Code — для написания и выполнения кода Python. Markdown — для текста, описаний, формул, заголовков.

Создание: Нажать кнопку + в панели сверху — добавится новая ячейка. Изменить тип можно в выпадающем меню (Code / Markdown) или горячими клавишами: Y — сделать ячейку кодом. M — сделать ячейку текстовой (Markdown).

Горячие клавиши работы в Jupyter Notebook, запуск на выполнение Основные сочетания: Ctrl + Enter — запустить текущую ячейку. Shift + Enter — запустить текущую ячейку и перейти к следующей. Alt + Enter — запустить и создать новую ячейку ниже. A — вставить ячейку выше. B — вставить ячейку ниже. D, D — удалить ячейку (нажать D дважды).

Режим редактирования и командный режим. Особенности Режим редактирования (Edit mode) — когда ты внутри ячейки и можешь печатать код или текст. Активируется клавишей Enter, ячейка подсвечивается зелёным. Командный режим (Command mode) — когда ты управляешь ячейками (добавляешь, удаляешь, меняешь тип). Активируется клавишей Esc, ячейка подсвечивается синим.

Решение задачи A. Математические действия

|Математическое действие        |Оператор в Python|	Пример    |
|-------------------------------|-----------------|-----------|
|Сложить	                    |+	              |2 + 3 = 5  |
|Вычесть	                    |-	              |7 - 4 = 3  |
|Умножить	                    |*	              |5 * 2 = 10 |
|Разделить	                    |/	              |8 / 2 = 4.0|
|Получить целую часть от деления|//               |	7 // 3 = 2|
|Получить остаток от деления	|%	              |7 % 3 = 1  |
|Возвести в степень	            |**	              |2 ** 3 = 8 |

Переменные. Имена. Правила создания Переменная — Ячейка памяти, в которую ложится значение. Например: x = flase pi = 3.12 name = "ArchLinux"

Правила: Имя начинается с буквы или подчёркивания (_), но не с цифры. Можно использовать только буквы, цифры и подчёркивание. Python регистрозависимый язык (т.е.Age ≠ age). Не использовать зарезервированные слова (if, for, while и т.п.).

Типы int, float и str int — целые числа (1, 25, -4) float — дробные числа (3.14, -0.5) str — строки, текст ("Привет", '123')

Приведение типов Иногда нужно преобразовать одно значение в другой тип: x = 5 # int y = float(x) # стало 5.0 z = str(x) # стало "5" n = int("12") # стало числом 12

Это называется приведение типов. Функции: int(), float(), str().

Логические операции

|Условие	      |Оператор в Python   | Пример                    |
|-----------------|--------------------|---------------------------|
|больше	          | >	               | 5 > 3 True                |
|больше или равно |	>=	               | 5 >= 5 True               |
|меньше	          | <	               | 2 < 7 True                |
|меньше или равно | <=	               | 3 <= 3 True               |
|не равно	      | !=	               | 4 != 5 True               |
|равно	          | ==	               | 2 == 2 True               |
|А и В	          | and	               | (5 > 2) and (3 < 4) True  |
|А или В	      | or	               | (5 > 6) or (2 < 3) True   |
